```python
def tuka_split(df_list):
    new_df_list = []
    for df in df_list:
        one = []
        two = []
        three = []
        four = []
        for i in df['通過']:
            tmp = i.split('-')
            if len(tmp) >= 3:
                three.append(tmp[2])
                if len(tmp) == 4:
                    four.append(tmp[3])
            one.append(tmp[0])
            two.append(tmp[1])
            if len(one) > len(three):
                three.append('-')
            elif len(one) > len(four):
                four.append('-')
        corner_rank = {'1コーナー':one,'2コーナー':two,'3コーナー':three,'4コーナー':four}
        if corner_rank['3コーナー'] == []:
            del corner_rank['4コーナー'], corner_rank['3コーナー']
        elif corner_rank['3コーナー'] != [] and corner_rank['4コーナー'] == []:
            del corner_rank['4コーナー']
        corner_df = pd.DataFrame(corner_rank)
        tmp = df.drop('通過',axis=1)
        new_df = pd.concat([tmp,corner_df],axis = 1)
        new_df_list.append(new_df)
    return new_df_list
```

```python
#計算用のデータ形成,エクセル出力に使用
df_list_all = [df_vic,df_in_3,df_ana]
df_list_ten = [df_vic_ten,df_in_3_ten,df_ana_ten]


#計算に使用
new_df_list_all = tuka_split(df_list_all)
new_df_list_ten = tuka_split(df_list_ten)
```

```python
def round_weight(df_list):
    n_df_list = []
    for df in df_list:
        new_weight = []
        for count,content in enumerate(df['馬体重']):
            new_weight.append(content-int(str(content)[-1]))
        df['馬体重'] = new_weight
        n_df_list.append(pd.DataFrame(df))
    return df_list
```

```python
def rate_cal(big_list,df_list):
    ans=[]
    for count,dic in enumerate(big_list):
        for count_,column in enumerate(dic):
            content_l = []
            list_ = []
            for count__,content in enumerate(dic[column]):
                df = df_list[count][column]
                bool_list = content == df
                if len(df[bool_list])/len(df) != 0:
                    content_l.append(content)
                    list_.append(len(df[bool_list])/len(df))
            if count_ == 0:
                df_ = pd.DataFrame({column:content_l,'':list_})
            else:
                df_ = pd.concat([df_,pd.DataFrame({column:content_l,'':list_})],axis=1).fillna('')
        ans.append(df_)
    return(ans)    
```

```python
#割合がまとめてある    0に勝ち馬のデータ、１に三着以内のデータ、２に穴馬のデータ
rate_ten = rate_cal(big_list_ten,n_df_list_ten)
rate_all = rate_cal(big_list_all,n_df_all)
```
